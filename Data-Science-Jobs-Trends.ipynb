## <a id="about-the-dataset"></a>2. About the Dataset


The data used in this project has been imported from Kaggle. which was retrieved from https://ai-jobs.net/ that collects information from employers and professionals all around the world. The data is available under an open-source license and we adhere to all the terms of usage of the data. The data is in .csv format comprising of 12 columns and 9,355 rows. It spans from 2020 â€“ 2023 years and includes salary information for approximately 100 job titles such as Data scientist, Data engineer, Machine Learning Engineer etc. These job titles have been further categorized into 10 job categories such as Data Science and Research, Machine Learning and AI etc. for easier analysis and visualization. 

The dataset includes following columns:

|Columns       | Description                                                                                                                                               |
|-----------------------|-----------------------------------------------------------------------------------------------------------------------------------------------------------|
| **work_year**          | The year in which the data was recorded. This field indicates the temporal context of the data, important for understanding salary trends over time.       |

| **salary_in_usd**      | The annual gross salary converted to United States Dollars (USD). This uniform currency conversion aids in global salary comparisons and analyses.         |
| **employee_residence** | The country of residence of the employee. This data point can be used to explore geographical salary differences and cost-of-living variations.             |
| **experience_level**   | Classifies the professional experience level of the employee. Common categories might include 'Entry-level', 'Mid-level', 'Senior', and 'Executive', providing insight into how experience influences salary in data-related roles. |
| **employment_type**    | Specifies the type of employment, such as 'Full-time', 'Part-time', 'Contract', etc. This helps in analyzing how different employment arrangements affect salary structures. |
| **work_setting**       | The work setting or environment, like 'Remote', 'In-person', or 'Hybrid'. This column reflects the impact of work settings on salary levels in the data industry. |
| **company_location**   | The country where the company is located. It helps in analyzing how the location of the company affects salary structures.                                 |
| **company_size**       | The size of the employer company, often categorized into small (S), medium (M), and large (L) sizes. This allows for analysis of how company size influences salary. |

### <a id="trends-for-data-science-jobs"></a>3.1 What are the trends for Data Science jobs year-per-year?

```python
plt.figure(figsize=(10, 6))

#Linegraph for average salary trends per experience level
sns.lineplot(data=jobs_df, x='year', y='salary_in_usd', hue='experience_level', estimator='mean')

plt.title('Average Data Science Salary by Experience Level - Year-wise', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Salary - USD', fontsize=12)
plt.legend(title='Experience Level', bbox_to_anchor=(1.05, 1), loc='upper left',
           title_fontsize='13', fontsize='11', fancybox=True, shadow=True, borderpad=1.2)

plt.xticks(ticks=jobs_df['year'].unique())

plt.show()
```

#### Reason of the overlap between the confidence interval / error bars across different experience levels (*feedback*)

### <a id="job-categories-salary"></a>3.2 How do job categories compare in salary?

As we enter the job market in Canada and USA, understanding the salary dynamics for jobs that exist for data professionals across different categories will help answer questions in regards to what are the hottest fields to go into, which fields yield the highest salaries?. In the dataset, the job titles are partitioned among the following 10 Job Categories: Data Science and Research, BI and Visualization, Data Architecture and Modeling, Data Analysis, Data Engineering, Leadership and Management, Data Quality and Operations, Machine Learning and AI, Cloud and Database, Data Management and Strategy. We took averages of salaries across different categories while focusing only on Canada and USA to keep the analysis relevant. A closer inspection by visaulizing most frequent job titiles by experience level can inform an aspiring data professional for best suited roles to apply for in order to attain a favorable job.  

```python
# Average salaries by job categories for USA
usa_jobs_df = jobs_df[jobs_df['country']=='United States']

avg_salary_usa = usa_jobs_df.groupby('job_category').agg(
    mean_salary=('salary_in_usd', 'mean')
).reset_index()
```

```python
# Average salaries by job categories for Canada

cad_jobs_df = jobs_df[jobs_df['country']=='Canada']

avg_salary_cad = cad_jobs_df.groupby('job_category').agg(
    mean_salary=('salary_in_usd', 'mean')
).reset_index()
```

```python
# Sorting data for salaries in descending order for plotting

sort_avg_salary_usa = avg_salary_usa.sort_values(by='mean_salary',ascending=False)
sort_avg_salary_cad = avg_salary_cad.sort_values(by='mean_salary',ascending=False)
```

```python
plt.figure(figsize=(20,8))

# Plot 1: USA salaries for job categories plot
plt.subplot(1,2,1)
plt.barh(sort_avg_salary_usa['job_category'], sort_avg_salary_usa['mean_salary'],
         color=plt.cm.Blues(np.linspace(1,0.1, len(sort_avg_salary_usa))))
plt.xlabel('Average Salaries (USD)')
plt.ylabel('Job Categories')
plt.title('Average Salaries across Job Categories in USA')
[plt.gca().invert_yaxis()]

# Plot 2: USA salaries for job categories plot
plt.subplot(1,2,2)
plt.barh(sort_avg_salary_cad ['job_category'], sort_avg_salary_cad ['mean_salary'],
         color=plt.cm.Reds(np.linspace(1, 0.1, len(sort_avg_salary_cad ))))
plt.xlabel('Average Salaries (USD)')
plt.ylabel('Job Categories')
plt.title('Average Salaries across Job Categories in Canada')
[plt.gca().invert_yaxis()]

plt.tight_layout()
plt.show()
```

**Detailed Analysis**

- In the USA, **Machine Learning and AI** positions get the highest average salaries, reflecting the growing demand for AI expertise.
- **Data Science and Research** and **Data Engineering** roles follow closely in terms of salary, as these are essential for both innovation and maintaining core data infrastructure.
- In Canada, while similar trends are observed, **Leadership and Management** roles have a stronger presence in terms of salary, indicating that Canadian companies are placing a significant emphasis on experienced leadership in data-driven strategies.
- The **Data Management and Strategy** category in Canada also shows higher-than-expected salaries, suggesting that strategic oversight is a key focus for Canadian organizations.
- There is a significant gap in salary between the USA and Canada, with Canadian companies offering lower average salaries in most categories, potentially due to differences in market dynamics, cost of living, and organizational budgets.

**Key Insights**

- In USA, technical roles like **Machine Learning and AI**, **Data Science**, and **Data Engineering** dominate the salary landscape, with a clear emphasis on advanced technical skills.
- In Canada, while technical roles remain important, there is a noticeable emphasis on **Leadership and Management** roles, which are compensated more competitively than in the USA.
- **Salary Gap**: Canada generally offers lower salaries compared to the USA across most categories, except for certain leadership positions, where compensation is higher in Canada.
- **Strategic Focus**: Canadian organizations appear to value strategic and leadership roles more, as compared to the USAâ€™s focus on cutting-edge technical expertise.

```python
# Graph to show which company location has how many job titles

# Grouping the data by 'country' and calculating the number of unique job titles for each country
job_titles_per_company_location = jobs_df.groupby('country')['job_title'].nunique().reset_index(name='job_title_count')

# sorting by the number of job titles and selecting the top 5 countries with highest jo title count
job_titles_per_company_location_sorted = job_titles_per_company_location.sort_values(by='job_title_count', ascending=False).head(5)

plt.figure(figsize=(8, 4))
sns.set_style('whitegrid')

# horizontal bar plot to visualize the countries with highest number of job titles
plt.barh(job_titles_per_company_location_sorted['country'],
         job_titles_per_company_location_sorted['job_title_count'],
         color='lightblue')

# adding labels and title
plt.xlabel('Number of Job Titles', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.title('Number of Job Titles per Company Location (Top 5 Countries)', fontsize=14)


plt.tight_layout()
plt.show()
```

```python
# graph to show how company size affects salary
from matplotlib import gridspec
from matplotlib.patches import Patch

# Group the dataframe by company_size and retrieve salary and experience_level columns
company_size_salary=jobs_df.groupby('company_size')[['salary_in_usd','experience_level']].apply(lambda x: x).reset_index()
company_size_salary.drop('level_1',axis=1,inplace=True)

# Sort the company_size values in descending order for better visualization
company_size_salary.sort_values('company_size',ascending=False,inplace=True)
company_size_salary.columns = ['company_size', 'salaries', 'experience_level']

plt.tight_layout()
plt.figure(figsize=(8,6))

# Plot: Boxplot to show salary distribution across company sizes
sns.boxplot(x=company_size_salary['company_size'],y=company_size_salary['salaries'])
plt.xlabel('Company Size')
plt.ylabel('Salary')
plt.title('Company size affecting Salaries')

# Custom legend patches
legend_elements = [
    Patch(facecolor='blue', edgecolor='blue', label='Small (0 - 50 employees)'),
    Patch(facecolor='green', edgecolor='green', label='Medium (50 - 500 employees)'),
    Patch(facecolor='red', edgecolor='red', label='Large (500+ employees)')
]

plt.legend(handles=legend_elements, title='Company Size Categories', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
```

**Detailed Analysis:**

The above box plot shows how company size affects the data science salaries.

The reason we are using boxplot here is because it provides a summary of the distribution through key statistics: the median, quartiles (Q1, Q3), and outliers.

When I say company size that means each company is categorized into three categories i.e Small, medium and large level. The company size here is categorized based on the number of employees in the company.



*   0 - 50 number of employees (Small company usually startups etc.)
*   50 - 500 number of employees (Medium level companies)
*   500 + number of employees (Large level companies)


**Key Insights**


1.   It is hypothesized that the larger the company, the higher they pay the data science professionals. However, the medium sized companies offer higher salaries than the large companies based on median data science salary against company size. Alongside, the number of outliers in the medium-sized companies is much higher than other company size and this could have contributes a lot to the distribution of the salaries.
2.   The salary distribution also becomes wider as the company size increases, indicating that large companies have a broader salary range, with more employees earning higher salaries. Smaller companies display a narrower salary range, clustered closer to the median. This may indicate that smaller firms have a more uniform pay structure.

<br>

```python
# Plot 2: Boxplot to show salary distribution across company sizes, with a hue to differentiate experience levels

plt.tight_layout()
plt.figure(figsize=(8,6))

sns.boxplot(x=company_size_salary['company_size'],y=company_size_salary['salaries'],hue=company_size_salary['experience_level'])
plt.xlabel('Company Size')
plt.ylabel('Salary')
plt.title('Company size per experience level affecting Salaries')
```

The above box plot shows how company size affects the data science salaries hued my experience level (Entry-level, Senior, Mid-level, Executive).

```python
# graph to show how employment type affects salary structures

# Grouping the dataset by employment type to analyze how salaries and work setting vary
employment_type_salary=pd.DataFrame(jobs_df.groupby('employment_type')[['salary_in_usd','work_setting']].apply(lambda x: x).reset_index())
employment_type_salary.drop('level_1',axis=1,inplace=True)
employment_type_salary.columns = ['employment_type', 'salaries', 'work_setting']

# removing row when employment_type is freelance and work_setting is In-person
# This condition is logically invalid as freelancers typically work remotely or in hybrid setups
employment_type_salary = employment_type_salary[
    ~((employment_type_salary['employment_type'] == 'Freelance') &
      (employment_type_salary['work_setting'] == 'In-person'))
]

plt.tight_layout()
plt.figure(figsize=(8,6))

# Plot: Boxplot to show how salaries are distributed across different employment types
sns.boxplot(x=employment_type_salary['employment_type'],y=employment_type_salary['salaries'])
plt.xlabel('Employment Type')
plt.ylabel('Salary')
plt.title('Employment Type affecting Salaries')

plt.subplots_adjust(wspace=0.5)
```

**Detailed Analysis**

The above box plot shows how Employment type affects the data science salaries.
So the Employment type is categorized in four categories:

*   Contract (Employee working on contract basis)
*   Freelance (Employee working as a freelancer)
*   Full-time (Employee working on full time basis)
*   Part-time (Employee working on part time basis)

**Key Insights**

- Full time employment has the widest salary range, with the highest salaries and some significant outliers on the upper end. This is expected, as full time positions often come with the most comprehensive compensation packages.

- Freelance and contract workers show a more concentrated range of salaries, significantly lower than full time roles. This could be due to the nature of short term or project based work.

- Part time roles tend to offer the lowest salaries, and their range is much narrower. This could be due to reduced hours or fewer responsibilities.

```python
# 2. Then for every country, find where they have different countries

foreignWorkersLst = []
for index, row in jobs_dfQ4.iterrows():
    # And I did notice that some people lived in different areas, but were not 'remote'
    if (row['country'] != row['employee_residence']) & (row['work_setting'] == 'Remote'):
        foreignWorkersLst.append(row)
foreignWorkers = pd.DataFrame(foreignWorkersLst, columns=jobs_dfQ4.columns)
display(foreignWorkers)

# Now we have a list of all the remote workers that work in different countries
```

*Note: the following is a calculation of the national averages. This means; take all the jobs provided by a given country, and find the average. This DOES NOT EXCLUDE remote workers who's countries differ, they are also included.*

```python
# lets calculate the mean of each country a company is located in

# Lets remove all the outsourced individuals to get a clean dataset for this analysis
jobs_dfQ4_CompanyLocation = jobs_dfQ4.groupby(by="country")
display(jobs_dfQ4_CompanyLocation)
avg_sal_country = jobs_dfQ4_CompanyLocation['salary_in_usd'].mean()
avg_sal_country.index.name = 'Country'
avg_sal_country.name = 'Average Income by Company Location'
display(avg_sal_country)
```

<br>

**The logic behind this code:**

- Get all the remote jobs.
- Filter those whose country of residence and company country differ.
- If they make less than the average, they are known as "outsourced workers" â€” the company outsources data-science work to reduce costs.
- If they make more than the average, they are known as "vacationer workers" â€” this can be justified if they were promoted to senior roles, or their skills are highly sought-after in the market, which leads to them being paid more than the company's national average.
- We will include error handling for cases where foreign employee pay equals the national average.

```python
# Let's see whether or not a given individual makes more or less than the average than they make in their company's country

outsourced = []
vacationer = []
samePay = []
for index, row in foreignWorkers.iterrows():
  # Include a buffer; standard deviation or 20% above
    if row['salary_in_usd'] < avg_sal_country.loc[row['country']]:
        outsourced.append(row)
    elif row['salary_in_usd'] > avg_sal_country.loc[row['country']]:
        vacationer.append(row)
    else:
        samePay.append(row)

outsourced = pd.DataFrame(outsourced, columns=jobs_dfQ4.columns)
vacationer = pd.DataFrame(vacationer, columns=jobs_dfQ4.columns)
samePay = pd.DataFrame(samePay, columns=jobs_dfQ4.columns)
print("outsourced: ")
display(outsourced)
print("vacationer: ")
display(vacationer)
print("same-pay: ")
display(samePay)
```

```python
# Now let's plot the graph to show the foreign worker's pay vs. average pay per company_location

# Calculate the average salary for outsourced workers by country and display the result
outsourcedMean = outsourced.groupby('country')['salary_in_usd'].mean()
outsourcedMean.index.name = 'Country' 
outsourcedMean.name = 'Outsourced Individual Income' 
display(outsourcedMean)  # Display the average salary of outsourced workers


# Create a DataFrame by joining the average salary of foreign workers with the average salary by country
jobs_df_avgforeignersSalary = avg_sal_country.to_frame().join(outsourcedMean.to_frame())


# Calculate the average salary for vacationer workers by country and display the result
vacationerMean = vacationer.groupby('country')['salary_in_usd'].mean()
vacationerMean.index.name = 'Country' 
vacationerMean.name = 'Vacationer Individual Income'
display(vacationerMean)  # Display the average salary of vacationer workers


# Join the average vacationer income to the existing DataFrame
jobs_df_avgforeignersSalary = jobs_df_avgforeignersSalary.join(vacationerMean.to_frame())


# Calculate the average salary for workers receiving the same pay as the national average
samePayMean = samePay.groupby('country')['salary_in_usd'].mean()
samePayMean.index.name = 'Country' 
samePayMean.name = 'Same-Pay Individual Income'
display(samePayMean)  # Display the average salary of workers receiving the same pay


# Join the same-pay individual income to the existing DataFrame with foreign worker salary information
jobs_df_avgforeignersSalary = jobs_df_avgforeignersSalary.join(samePayMean.to_frame())
display(jobs_df_avgforeignersSalary)

print("Outsourced")
```

Entry-level and mid-level roles have seen consistent salary growth, signifying that the data science job market is not saturated and there is an increasing demand for new talent and experienced professionals. However, senior and executive-level salaries showed some fluctuations, potentially due to market saturation or changing demand at the top levels.

Additionally, salary analysis across job categories in the USA and Canada highlighted that technical roles like Machine Learning and AI command the highest salaries in both countries, while leadership roles are more prominent in Canada. The overall salary gap between the two countries suggests differences in market maturity and organizational priorities. 

When examining company size and employment type, medium-sized companies tend to offer competitive salaries, especially for senior roles, whereas full-time employment offers the widest salary range. Remote and contract work are becoming more popular, particularly for freelance professionals, as remote settings offer higher salaries for specific roles.

The increasing ratio of average data science salaries to minimum wages over the years suggests a growing demand and higher compensation in this field. However, a high ratio may also indicate low minimum wages in certain countries, which can distort salary comparisons. This analysis provides valuable insights into which countries may be more favorable for pursuing a career in data science, highlighting areas where professionals can expect competitive salaries relative to living costs.
